```python
def add(a,b):
  return a+b
def sub(a,b):
  return a-b
def mul(a,b):
  return a*b
def div(a,b):
  return a/b
def mod(a,b):
  return a%b
def floor(a,b):
  return a//b
a=int(input())
b=int(input())
print(add(a,b))
print(sub(a,b))
print(mul(a,b))
print(div(a,b))
print(mod(a,b))
print(floor(a,b))
```

```python
#random
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(5)
y=np.random.rand(5)
plt.plot(x,y,marker='o')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#scatter plot using random
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(5)
y=np.random.rand(5)
plt.scatter(x,y,marker='o')
plt.title("scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.bar(x,y)
plt.title("Bar Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
